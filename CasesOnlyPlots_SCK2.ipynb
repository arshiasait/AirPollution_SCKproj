```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime as dt
import datetime
import csv
import math
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from coordinateSystems import GeographicSystem
from coordinateSystems import TangentPlaneCartesianSystem
```

```python
SCK2_ALL_cases = pd.read_csv('ALL_SCK2_cases.csv')
SCK2_ALL_nocases = pd.read_csv('ALL_SCK2_nocases.csv')
SCK2_ten_cases = pd.read_csv('ten_SCK2_cases.csv')
SCK2_ten_nocases = pd.read_csv('ten_SCK2_nocases.csv')
```

```python
def calcs(data):
    print 'mean:',np.mean(data)
    print 'stdev:',np.std(data)
```

```python
calcs(SCK2_ALL_cases['SCK2_CO'])
```

```python
calcs(SCK2_ALL_nocases['SCK2_CO'])
```

```python
calcs(SCK2_ten_cases['SCK2_CO'])
```

```python
calcs(SCK2_ten_nocases['SCK2_CO'])
```

```python
fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(12,1,(1,8))
ax.set_xlim(-990,1390)
ax.set_ylim(-800,1000)
ax.scatter(0,0,c='black',s=100,linewidths=2,marker='x') # point of fixed site
ax.scatter(SCK2_ten_cases['X(m)'],SCK2_ten_cases['Y(m)'],s=2,c='r',linewidth=0)

ax.set_title('SCK2 Carbon Monoxide (K$\Omega$)',fontsize=16)
ax.set_xlabel('East-West (meters)',fontsize=16)
ax.set_ylabel('North-South (meters)',fontsize=16)
majorLocator = MultipleLocator(200)
majorFormatter = FormatStrFormatter('%d')
minorLocator = MultipleLocator(100)
ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_major_formatter(majorFormatter)
ax.xaxis.set_minor_locator(minorLocator)
ax.tick_params(labelsize=14)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_major_formatter(majorFormatter)
ax.yaxis.set_minor_locator(minorLocator)
ax.grid()
#plt.savefig('ALL_CO_ten_SCK2', format='png', dpi=300)
```
